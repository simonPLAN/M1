Dans le programme ci-dessous on a :
- Une création de modèle de classificateur bayésien de tailles de vecteurs diverses : 1/2/3/4 champs contenus dans les données d'origine à partir des 100 premières données qui constituent notre corpus d'apprentissage

- une application de ces modèles sur un corpus de développement des 30 valeurs suivantes afin de déterminer le meilleur modèle pour classifier des données d'Iris

- on remarque que le meilleur modèle est constitué d'un vecteur 1 avec comme champs la largeur de la pétale avec un taux d'erreur sur le corpus de développement d'environ 6%

- l'application de ce meilleur modèle sur le corpus de test contenant les 20 dernières valeurs de nos données où on observe un taux d'erreur final de 10%

```python
from sklearn import datasets
import numpy as np
import scipy

LONGUEUR_SEPALE = 0
LARGEUR_SEPALE = 1
LONGUEUR_PETALE = 2
LARGEUR_PETALE = 3

# Calcul la moyenne des valeurs de la cible dans les données d'apprentissage
def calcul_moyenne(data, target, classe, param, nb_espece):
    compteur = 0
    for i in range(100) :
        if (target[i] == classe) :
            compteur += data[i][param]
    return compteur/nb_espece

# Créer un vecteur 1 à partir des données d'apprentissage en fonction de param1
def etablir_vecteur_1(data, target, classe, param1):
    vecteur = []
    for i in range(100) :
        if (target[i] == classe) :
            vecteur.append(data[i][param1])
    return vecteur

# Créer un vecteur 2 à partir des données d'apprentissage en fonction des param1 et param2
def etablir_vecteur_2(data, target, classe, param1, param2):
    vecteur = []
    for i in range(100) :
        if (target[i] == classe) :
            vecteur.append([data[i][param1], data[i][param2]])
    return vecteur
    
# Créer un vecteur 3 à partir des données d'apprentissage en fonction des param1 et param2 et param3
def etablir_vecteur_3(data, target, classe, param1, param2, param3):
    vecteur = []
    for i in range(100) :
        if (target[i] == classe) :
            vecteur.append([data[i][param1], data[i][param2], data[i][param3]])
    return vecteur

# Créer un vecteur 4 à partir des données d'apprentissage en fonction des param1 et param2 et param3 et param4
def etablir_vecteur_4(data, target, classe, param1, param2, param3, param4):
    vecteur = []
    for i in range(100) :
        if (target[i] == classe) :
            vecteur.append([data[i][param1], data[i][param2], data[i][param3], data[i][param4]])
    return vecteur

# Calcul de la loi Gaussienne
def calcul_Gauss_simple(x, moy, var):
    return scipy.stats.norm.pdf(x,moy,var)

# Calcul de la probabilité d'un vecteur 1
def calcul_Proba_Vec1(x1, moy1, var1, p_priori):
    return (calcul_Gauss_simple(x1, moy1, var1) * p_priori)
                            
def generation_Modele_1(target1, data, target, p_priori_0, p_priori_1, p_priori_2):
    
    # On constitue le corpus d'apprentissage avec les 100 premières données du corpus
    corpus_apprentissage_data = data[0:100]
    corpus_apprentissage_target = target[0:100]
    
    # On créer des vecteurs 2 sur les paramètres voulues en les triant par Iris
    vec_0 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 0, target1)
    vec_1 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 1, target1)
    vec_2 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 2, target1)

    # Moyenne des sous corpus triés par type
    moy_0 = []
    moy_0.append(np.mean(vec_0,0))
    moy_1 = []
    moy_1.append(np.mean(vec_1,0))
    moy_2 = []
    moy_2.append(np.mean(vec_2,0))
    

    # Variance des sous corpus triés par type
    var_0 = []
    var_0.append(np.std(vec_0,0))
    var_1 = []
    var_1.append(np.std(vec_1,0))
    var_2 = []
    var_2.append(np.std(vec_2,0))

    #print("---------------------------------------------")

    tri_corpus_dev_0 = []
    tri_corpus_dev_1 = []
    tri_corpus_dev_2 = []
    tri_corpus_dev_target = []
    corpus_dev_data = data[100:130]
    corpus_dev_target = target[100:130]
    
    compteur_dev_0 = compteur_dev_1 = compteur_dev_2 = 0
    nb_erreur = 0

    for i in corpus_dev_target :
        if i == 0 : 
            compteur_dev_0 += 1
        elif i == 1: 
            compteur_dev_1 += 1
        elif i == 2: 
            compteur_dev_2 += 1
        else :
            print ("Erreur, espece non repertorie")

    for i in range(len(corpus_dev_data)):
        proba_0 = calcul_Proba_Vec1(corpus_dev_data[i][target1], moy_0[0], var_0[0], p_priori_0)
        proba_1 = calcul_Proba_Vec1(corpus_dev_data[i][target1], moy_1[0], var_1[0], p_priori_1)
        proba_2 = calcul_Proba_Vec1(corpus_dev_data[i][target1], moy_2[0], var_2[0], p_priori_2)
        if (proba_0 > proba_1 and proba_0 > proba_2):
            tri_corpus_dev_0.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(0)
        elif (proba_1 > proba_0 and proba_1 > proba_2):
            tri_corpus_dev_1.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(1)
        else:
            tri_corpus_dev_2.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(2)

    for i in range(len(tri_corpus_dev_target)) :
        if tri_corpus_dev_target[i] != corpus_dev_target[i] :
            nb_erreur += 1

    print("Taux d'erreur du modèle de vecteur 1 comme champs : ", target1, " = ", (nb_erreur/30)*100, "%")

    #print("---------------------FIN GENERATION VECTEUR 1------------------------")
                            
# Calcul de la probabilité d'un vecteur 2
def calcul_Proba_Vec2(x1, x2, moy1, moy2, var1, var2, p_priori):
    return (calcul_Gauss_simple(x1, moy1, var1) * calcul_Gauss_simple(x2, moy2, var2) * p_priori)

def generation_Modele_2(target1, target2, data, target, p_priori_0, p_priori_1, p_priori_2):
    
    # On constitue le corpus d'apprentissage avec les 100 premières données du corpus
    corpus_apprentissage_data = data[0:100]
    corpus_apprentissage_target = target[0:100]
    
    # On créer des vecteurs 2 sur les paramètres voulues en les triant par Iris
    vec_0 = etablir_vecteur_2(corpus_apprentissage_data, corpus_apprentissage_target, 0, target1, target2)
    vec_1 = etablir_vecteur_2(corpus_apprentissage_data, corpus_apprentissage_target, 1, target1, target2)
    vec_2 = etablir_vecteur_2(corpus_apprentissage_data, corpus_apprentissage_target, 2, target1, target2)

    # Moyenne des sous corpus triés par type
    moy_0 = []
    moy_0.append(np.mean(vec_0,0))
    moy_1 = []
    moy_1.append(np.mean(vec_1,0))
    moy_2 = []
    moy_2.append(np.mean(vec_2,0))

    # Variance des sous corpus triés par type
    var_0 = []
    var_0.append(np.std(vec_0,0))
    var_1 = []
    var_1.append(np.std(vec_1,0))
    var_2 = []
    var_2.append(np.std(vec_2,0))
                            
    tri_corpus_dev_0 = []
    tri_corpus_dev_1 = []
    tri_corpus_dev_2 = []
    tri_corpus_dev_target = []
    corpus_dev_data = data[100:130]
    corpus_dev_target = target[100:130]
    
    compteur_dev_0 = compteur_dev_1 = compteur_dev_2 = 0
    nb_erreur = 0

    for i in corpus_dev_target :
        if i == 0 : 
            compteur_dev_0 += 1
        elif i == 1: 
            compteur_dev_1 += 1
        elif i == 2: 
            compteur_dev_2 += 1
        else :
            print ("Erreur, espece non repertorie")

    for i in range(len(corpus_dev_data)):
        proba_0 = calcul_Proba_Vec2(corpus_dev_data[i][target1], corpus_dev_data[i][target2], moy_0[0][0], moy_0[0][1], var_0[0][0], var_0[0][1], p_priori_0)
        proba_1 = calcul_Proba_Vec2(corpus_dev_data[i][target1], corpus_dev_data[i][target2], moy_1[0][0], moy_1[0][1], var_1[0][0], var_1[0][1], p_priori_1)
        proba_2 = calcul_Proba_Vec2(corpus_dev_data[i][target1], corpus_dev_data[i][target2], moy_2[0][0], moy_2[0][1], var_2[0][0], var_2[0][1], p_priori_2)
        if (proba_0 > proba_1 and proba_0 > proba_2):
            tri_corpus_dev_0.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(0)
        elif (proba_1 > proba_0 and proba_1 > proba_2):
            tri_corpus_dev_1.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(1)
        else:
            tri_corpus_dev_2.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(2)

    for i in range(len(tri_corpus_dev_target)) :
        if tri_corpus_dev_target[i] != corpus_dev_target[i] :
            nb_erreur += 1

    print("Taux d'erreur du modèle de vecteur 2 comme champs : ", target1, " et ", target2, " = ", (nb_erreur/30)*100, "%")

    #print("---------------------FIN GENERATION VECTEUR 2------------------------")
                            
# Calcul de la probabilité d'un vecteur 3
def calcul_Proba_Vec3(x1, x2, x3, moy1, moy2, moy3, var1, var2, var3, p_priori):
    return (calcul_Gauss_simple(x1, moy1, var1) * calcul_Gauss_simple(x2, moy2, var2) * calcul_Gauss_simple(x3, moy3, var3) * p_priori)

def generation_Modele_3(target1, target2, target3, data, target, p_priori_0, p_priori_1, p_priori_2):
    
    # On constitue le corpus d'apprentissage avec les 100 premières données du corpus
    corpus_apprentissage_data = data[0:100]
    corpus_apprentissage_target = target[0:100]
    
    # On créer des vecteurs 3 sur les paramètres voulues en les triant par Iris
    vec_0 = etablir_vecteur_3(corpus_apprentissage_data, corpus_apprentissage_target, 0, target1, target2, target3)
    vec_1 = etablir_vecteur_3(corpus_apprentissage_data, corpus_apprentissage_target, 1, target1, target2, target3)
    vec_2 = etablir_vecteur_3(corpus_apprentissage_data, corpus_apprentissage_target, 2, target1, target2, target3)

    # Moyenne des sous corpus triés par type
    moy_0 = []
    moy_0.append(np.mean(vec_0,0))
    moy_1 = []
    moy_1.append(np.mean(vec_1,0))
    moy_2 = []
    moy_2.append(np.mean(vec_2,0))

    # Variance des sous corpus triés par type
    var_0 = []
    var_0.append(np.std(vec_0,0))
    var_1 = []
    var_1.append(np.std(vec_1,0))
    var_2 = []
    var_2.append(np.std(vec_2,0))

    tri_corpus_dev_0 = []
    tri_corpus_dev_1 = []
    tri_corpus_dev_2 = []
    tri_corpus_dev_target = []
    corpus_dev_data = data[100:130]
    corpus_dev_target = target[100:130]
    
    compteur_dev_0 = compteur_dev_1 = compteur_dev_2 = 0
    nb_erreur = 0

    for i in corpus_dev_target :
        if i == 0 : 
            compteur_dev_0 += 1
        elif i == 1: 
            compteur_dev_1 += 1
        elif i == 2: 
            compteur_dev_2 += 1
        else :
            print ("Erreur, espece non repertorie")

    for i in range(len(corpus_dev_data)):
        proba_0 = calcul_Proba_Vec3(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], moy_0[0][0], moy_0[0][1], moy_0[0][2], var_0[0][0], var_0[0][1], var_0[0][2], p_priori_0)
        proba_1 = calcul_Proba_Vec3(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], moy_1[0][0], moy_1[0][1], moy_1[0][2], var_1[0][0], var_1[0][1], var_1[0][2], p_priori_1)
        proba_2 = calcul_Proba_Vec3(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], moy_2[0][0], moy_2[0][1], moy_2[0][2], var_2[0][0], var_2[0][1], var_2[0][2], p_priori_2)
        if (proba_0 > proba_1 and proba_0 > proba_2):
            tri_corpus_dev_0.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(0)
        elif (proba_1 > proba_0 and proba_1 > proba_2):
            tri_corpus_dev_1.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(1)
        else:
            tri_corpus_dev_2.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(2)

    for i in range(len(tri_corpus_dev_target)) :
        if tri_corpus_dev_target[i] != corpus_dev_target[i] :
            nb_erreur += 1

    print("Taux d'erreur du modèle de vecteur 3 comme champs : ", target1, " et ", target2, " et ", target3, " = ", (nb_erreur/30)*100, "%")

    #print("---------------------FIN GENERATION VECTEUR 3------------------------")
    
# Calcul de la probabilité d'un vecteur 4
def calcul_Proba_Vec4(x1, x2, x3, x4, moy1, moy2, moy3, moy4, var1, var2, var3, var4, p_priori):
    return (calcul_Gauss_simple(x1, moy1, var1) * calcul_Gauss_simple(x2, moy2, var2) * calcul_Gauss_simple(x3, moy3, var3) * calcul_Gauss_simple(x4, moy4, var4) *p_priori)

def generation_Modele_4(target1, target2, target3, target4, data, target, p_priori_0, p_priori_1, p_priori_2):
    
    # On constitue le corpus d'apprentissage avec les 100 premières données du corpus
    corpus_apprentissage_data = data[0:100]
    corpus_apprentissage_target = target[0:100]
    
    # On créer des vecteurs 4 sur les paramètres voulues en les triant par Iris
    vec_0 = etablir_vecteur_4(corpus_apprentissage_data, corpus_apprentissage_target, 0, target1, target2, target3, target4)
    vec_1 = etablir_vecteur_4(corpus_apprentissage_data, corpus_apprentissage_target, 1, target1, target2, target3, target4)
    vec_2 = etablir_vecteur_4(corpus_apprentissage_data, corpus_apprentissage_target, 2, target1, target2, target3, target4)

    # Moyenne des sous corpus triés par type
    moy_0 = []
    moy_0.append(np.mean(vec_0,0))
    moy_1 = []
    moy_1.append(np.mean(vec_1,0))
    moy_2 = []
    moy_2.append(np.mean(vec_2,0))

    # Variance des sous corpus triés par type
    var_0 = []
    var_0.append(np.std(vec_0,0))
    var_1 = []
    var_1.append(np.std(vec_1,0))
    var_2 = []
    var_2.append(np.std(vec_2,0))

    tri_corpus_dev_0 = []
    tri_corpus_dev_1 = []
    tri_corpus_dev_2 = []
    tri_corpus_dev_target = []
    corpus_dev_data = data[100:130]
    corpus_dev_target = target[100:130]
    
    compteur_dev_0 = compteur_dev_1 = compteur_dev_2 = 0
    nb_erreur = 0

    for i in corpus_dev_target :
        if i == 0 : 
            compteur_dev_0 += 1
        elif i == 1: 
            compteur_dev_1 += 1
        elif i == 2: 
            compteur_dev_2 += 1
        else :
            print ("Erreur, espece non repertorie")

    for i in range(len(corpus_dev_data)):
        proba_0 = calcul_Proba_Vec4(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], corpus_dev_data[i][target4], moy_0[0][0], moy_0[0][1], moy_0[0][2], moy_0[0][3], var_0[0][0], var_0[0][1], var_0[0][2], var_0[0][3], p_priori_0)
        proba_1 = calcul_Proba_Vec4(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], corpus_dev_data[i][target4], moy_1[0][0], moy_1[0][1], moy_1[0][2], moy_1[0][3], var_1[0][0], var_1[0][1], var_1[0][2], var_1[0][3], p_priori_1)
        proba_2 = calcul_Proba_Vec4(corpus_dev_data[i][target1], corpus_dev_data[i][target2], corpus_dev_data[i][target3], corpus_dev_data[i][target4], moy_2[0][0], moy_2[0][1], moy_2[0][2], moy_2[0][3], var_2[0][0], var_2[0][1], var_2[0][2], var_2[0][3], p_priori_2)
        if (proba_0 > proba_1 and proba_0 > proba_2):
            tri_corpus_dev_0.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(0)
        elif (proba_1 > proba_0 and proba_1 > proba_2):
            tri_corpus_dev_1.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(1)
        else:
            tri_corpus_dev_2.append(corpus_dev_data[i])
            tri_corpus_dev_target.append(2)

    for i in range(len(tri_corpus_dev_target)) :
        if tri_corpus_dev_target[i] != corpus_dev_target[i] :
            nb_erreur += 1

    print("Taux d'erreur du modèle de vecteur 4 comme champs : ", target1, " et ", target2, " et ", target3, " et", target4, " = ", (nb_erreur/30)*100, "%")

    #print("---------------------FIN GENERATION VECTEUR 4------------------------")
    
def application_corpus_test(target1, data, target, p_priori_0, p_priori_1, p_priori_2):
    # On constitue le corpus d'apprentissage avec les 100 premières données du corpus
    corpus_apprentissage_data = data[0:100]
    corpus_apprentissage_target = target[0:100]
    
    # On créer des vecteurs 2 sur les paramètres voulues en les triant par Iris
    vec_0 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 0, target1)
    vec_1 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 1, target1)
    vec_2 = etablir_vecteur_1(corpus_apprentissage_data, corpus_apprentissage_target, 2, target1)

    # Moyenne des sous corpus triés par type
    moy_0 = []
    moy_0.append(np.mean(vec_0,0))
    moy_1 = []
    moy_1.append(np.mean(vec_1,0))
    moy_2 = []
    moy_2.append(np.mean(vec_2,0))
    

    # Variance des sous corpus triés par type
    var_0 = []
    var_0.append(np.std(vec_0,0))
    var_1 = []
    var_1.append(np.std(vec_1,0))
    var_2 = []
    var_2.append(np.std(vec_2,0))

    #print("---------------------------------------------")

    tri_corpus_test_0 = []
    tri_corpus_test_1 = []
    tri_corpus_test_2 = []
    tri_corpus_test_target = []
    corpus_test_data = data[130:150]
    corpus_test_target = target[130:150]
    
    compteur_test_0 = compteur_test_1 = compteur_test_2 = 0
    nb_erreur = 0

    for i in corpus_test_target :
        if i == 0 : 
            compteur_test_0 += 1
        elif i == 1: 
            compteur_test_1 += 1
        elif i == 2: 
            compteur_test_2 += 1
        else :
            print ("Erreur, espece non repertorie")

    for i in range(len(corpus_test_data)):
        proba_0 = calcul_Proba_Vec1(corpus_test_data[i][target1], moy_0[0], var_0[0], p_priori_0)
        proba_1 = calcul_Proba_Vec1(corpus_test_data[i][target1], moy_1[0], var_1[0], p_priori_1)
        proba_2 = calcul_Proba_Vec1(corpus_test_data[i][target1], moy_2[0], var_2[0], p_priori_2)
        if (proba_0 > proba_1 and proba_0 > proba_2):
            tri_corpus_test_0.append(corpus_test_data[i])
            tri_corpus_test_target.append(0)
        elif (proba_1 > proba_0 and proba_1 > proba_2):
            tri_corpus_test_1.append(corpus_test_data[i])
            tri_corpus_test_target.append(1)
        else:
            tri_corpus_test_2.append(corpus_test_data[i])
            tri_corpus_test_target.append(2)

    for i in range(len(tri_corpus_test_target)) :
        if tri_corpus_test_target[i] != corpus_test_target[i] :
            nb_erreur += 1

    print("Taux d'erreur du modèle de vecteur 1 comme champs : ", target1, " = ", (nb_erreur/20)*100, "%")

    #print("---------------------FIN APPLICATION CORPUS TEST ------------------------")
    
iris = datasets.load_iris()

Ciris = np.c_[iris.data.reshape(len(iris.data), -1), iris.target.reshape(len(iris.target), -1)]
np.random.seed(987654321)
np.random.shuffle(Ciris)
shuffledIrisData = Ciris[ :, :iris.data.size//len(iris.data)].reshape(iris.data.shape)
shuffledIrisTarget = Ciris[ :, iris.data.size//len(iris.data) :].reshape(iris.target.shape)

# On constitue le corpus d'apprentissage avec les 100 premières données du corpus
corpus_apprentissage_data = shuffledIrisData[0:100]
corpus_apprentissage_target = shuffledIrisTarget[0:100]

# Compteur de chaque éléments de chaque classe dans le corpus d'apprentissage
compteur_0 = 0
compteur_1 = 0
compteur_2 = 0

for i in corpus_apprentissage_target :
    if i == 0 : 
        compteur_0 += 1
    elif i == 1: 
        compteur_1 += 1
    elif i == 2: 
        compteur_2 += 1
    else :
        print ("Erreur, espece non repertorie")

# Taille du corpus d'apprentissage
taille_corpus = compteur_0+compteur_1+compteur_2

# On calcul les propabilités à priori
p_priori_0 = compteur_0 / taille_corpus
p_priori_1 = compteur_1 / taille_corpus
p_priori_2 = compteur_2 / taille_corpus

print("------------------------")

print("---- CLASSIFICATION ----")

print("------------------------")
 
for i in range(4) :
    generation_Modele_1(i, shuffledIrisData, shuffledIrisTarget, p_priori_0, p_priori_1, p_priori_2)
    print("------------------------")
                            
for i in range(4) :
    for j in range (4) :
        if (i != j) :
            generation_Modele_2(i, j, shuffledIrisData, shuffledIrisTarget, p_priori_0, p_priori_1, p_priori_2)
            print("------------------------")


print("------------------------")
for i in range(4) :
    for j in range (4) :
        for k in range (4) :
            if (i != j and i != k and j != k) :
                generation_Modele_3(i, j, k, shuffledIrisData, shuffledIrisTarget, p_priori_0, p_priori_1, p_priori_2)
                print("------------------------")

generation_Modele_4(0, 1, 2, 3, shuffledIrisData, shuffledIrisTarget, p_priori_0, p_priori_1, p_priori_2)


print("----------------------------------------------------------")

print("---- APPLICATION DU MEILLEUR MODELE AU CORPUS DE TEST ----")

print("----------------------------------------------------------")

application_corpus_test(3, shuffledIrisData, shuffledIrisTarget, p_priori_0, p_priori_1, p_priori_2)

```
